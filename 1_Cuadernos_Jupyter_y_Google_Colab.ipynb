## <font color='steelblue'>1. Jupyter Notebooks y Google Colab</font>


### <font color='steelblue'>¿Qué son los Jupyter Notebooks?</font>

Un cuaderno Jupyter, o **Jupyter Notebook**, es una aplicación web desarrollada bajo el [Proyecto Jupyter](https://jupyter.org/), que proporciona un entorno web interactivo de código abierto para crear y compartir documentos. Estos documentos pueden contener secuencias de código de programación, resultados de ejecución (tablas, gráficos, etc.), y textos para explicar y comentar los resultados.

### <font color='steelblue'>Los componentes de Jupyter Notebooks</font>

Los [Jupyter Notebooks](https://jupyter.org/) combinan tres componentes:

* La **aplicación web**: una aplicación web interactiva para escribir y ejecutar código de manera interactiva.

* **Entorno de ejecución**: procesos separados iniciados por la aplicación web que ejecuta el código de los usuarios en un lenguaje de programación determinado y devuelve la salida a la aplicación web del cuaderno.

* **Documentos o cuadernos**, que se pueden generar auto-contenidos, que contienen una representación de todo el contenido visible en la aplicación web del cuaderno, incluidas las entradas y salidas de los cálculos, el texto narrativo, las ecuaciones, las imágenes y las representaciones multimedia de objetos. Cada documento o cuaderno tiene su propio entorno de ejecución.


### <font color='steelblue'>La Aplicación web Colaboratory en el entorno gratuito de Jupyter Notebook</font>

La aplicación web [COLABORATORY](https://colab.research.google.com/notebooks/welcome.ipynb), también conocida como **Google Colab**, permite a los usuarios de *Google*:

   * Editar el código en el navegador.

   * Ejecutar el código desde el navegador.

   * Visualizar los resultados de los cálculos junto al código que los generó.

   * Comentar los resultados utilizando el lenguaje de marcas *Markdown*.

Todos los cuadernos Colab se almacenan y gestionan respecto a edición y permisos, como cualquier otro documento de Google desde Google Drive.

## <font color='steelblue'>4. Insertar ecuaciones</font>


Si queremos insertar ecuaciones en las celdas de código, es preciso tener algunos conocimientos básicos sobre $\LaTeX$, pues es el lenguaje que se utiliza para ello.
Es interesante tener localizado cómo se escriben los símbolos más habituales, como nos proporcionan por ejemplo en esta [guía rápida o *CheatSheet*](http://users.dickinson.edu/~richesod/latex/latexcheatsheet.pdf).

Cuando mostremos una expresión LaTeX, se mostrará el Markdown/LaTeX en bruto debajo de ella.

### <font color='steelblue'>$\LaTeX$ básico</font>


Para escribir $\LaTeX$ dentro de un texto, hay que ubicarlo siempre entre signos de dólar (`$`). Todas las fórmulas en $\LaTeX$ se muestran en cursiva.

Como ejemplo, consideramos la siguiente oración y el código Markdown/LaTeX para representarlo.

```markdown
El Área del cuadrado es $Área=l^2$.
```
>El Área del cuadrado es $Área=l^2$.


Se ha utilizado `^2` para elevar al cuadrado; si se tratara de un subíndice, sería `_2`. Si la potencia (o el subíndice) tiene más de un caracter, debe ir entre llaves `{}`. De forma general, las llaves se utilizan en LaTeX para agrupar los símbolos contenidos en una expresión.

```markdown
Euler nos contó que $\mathrm{e}^{i \pi} + 1 = 0$.
```
>Euler nos contó que $\mathrm{e}^{i \pi} + 1 = 0$.






Además de las llaves de agrupación, en la anterior fórmula hay otros elementos sintácticos interesantes:

- En primer lugar, se ha convertido el carácter especial $\pi$ escribiendo `\pi`. En general, una barra invertida precede a los símbolos o comandos especiales en $\LaTeX$. Si queremos representar una letra griega como $\theta$, usamos `\theta`.

- También se ha utilizado `\mathrm{e}` para la base del logaritmo natural.  El comando `\mathrm{}` se utiliza cuando pretendemos usar fuente romana para todo el texto entre llaves, y no cursiva.

```markdown
Esto es $e$ con la fuente por defecto para símbolos. Esto es $\mathrm{e}$, con fuente romana.
```
>Esto es $e$ con la fuente por defecto para símbolos. Esto es $\mathrm{e}$, con fuente romana.

En la siguiente expresión, además de las llaves de agrupación para subíndices, se utiliza el comando `\mathbf{}` resaltar en negrita lo que está entre llaves:

```markdown
El producto escalar de dos  vectores de dimensión $n$ es $\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i$.
```

>El producto escalar de dos vectores de dimensión $n$ es $\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i$.

De igual modo, volvemos a utilizar a continuación las llaves para agrupar la expresión bajo la raíz cuadrada,

```markdown
\sqrt{3x-1}+(1+x)^2
```
> $\sqrt{3x-1}+(1+x)^2$

Y el operador `frac{}{}` para definir una fracción, junto con el operador `\` para insertar un espacio en blanco entre dos símbolos:

```markdown
e^x=\sum_{i=0}^\infty \frac{1}{i!} \ x^i$
```
>$e^x=\sum_{i=0}^\infty \frac{1}{i!}\ x^i$


### <font color='steelblue'>Centrar ecuaciones</font>

A veces, al escribir una ecuación deseamos escribirla aislada en una línea y centrada. Para ello podemos utilizar varias opciones:

- Utilizar el comando $\text{\begin{align} ... \end{align</font> y emplazar la fórmula en los puntos suspensivos.

```markdown
\begin{align}
\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i
\end{align}
```

>\begin{align}
\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i
\end{align}

- el comando $\text{\begin{equation}...\end{equation</font>

```markdown
\begin{equation}
\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i
\end{equation}
```

>\begin{equation}
\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i
\end{equation}

- Utilizar el doble dólar `$$` para iniciar y finalizar una fórmula.


```markdown
$$\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i $$
```
>$$\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i $$


Si queremos escribir una ecuación en dos líneas, hay que usar dos contra-barras `\\` para el salto:

```markdown
\begin{align}
\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i \\
= a_x b_x + a_y b_y + a_z b_z.
\end{align}
```

>\begin{align}
\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^n a_i b_i \\
= a_x b_x + a_y b_y + a_z b_z.
\end{align}

Si además queremos centrar las dos líneas en torno al símbolo de igualdad, basta utilizar el símbolo `&` delante del símbolo que pretendemos que sirva de centro:

```markdown
\begin{align}
\mathbf{a} \cdot \mathbf{b} &= \sum_{i=1}^n a_i b_i \\
&= a_x b_x + a_y b_y + a_z b_z.
\end{align}
```

>\begin{align}

### <font color='steelblue'>Matrices y funciones</font>

Para escribir matrices podemos utilizar el entorno `pmatrix`, con el comando `begin{}...\end{}`. Los elementos de una fila se separan mediante el símbolo `&` y las filas mediante una contra-barra `\\`.

Por ejemplo, una matriz de covarianza para dos variables puede escribirse como:


```markdown
$$\sigma^2 = \begin{pmatrix}
\sigma_1^2 & \sigma_{12} \\
\sigma_{12} & \sigma_2^2
\end{pmatrix}.$$
```

>$$
\sigma^2 = \begin{pmatrix}
\sigma_1^2 & \sigma_{12} \\
\sigma_{12} & \sigma_2^2
\end{pmatrix}.
$$

Para representar unos valores apilados sobre otros a modo de tabla, se puede utilizar el entorno `array`, especificando entre llaves y con las letras `c`, `l` o `r`, la alineación deseada para cada columna,  centradas, alineadas a la izquierda o a la derecha, respectivamente. Los delimitadores de la tabla se definen con `\left` y `\right`, seguidos del símbolo que proceda: llaves `\{`, corchetes `[` o paréntesis `(`

```markdown
\left[\begin{array}{cc}
a & b \\
c & d
\end{array}\right]
```

>$$  
\left[\begin{array}{cc}
a & b \\
c & d
\end{array}\right]$$

Cuando queremos definir una función definida a trozos, podemos utilizar el entorno `cases`:

```markdown
f(x)=\begin{cases}
0 & \text{ si } x \le 0 \\
x & \text{ si }  x > 0.
\end{cases}
```

>$$f(x)=
\begin{cases}
0 & \text{ si } x \le 0 \\
x & \text{ si } x > 0.
\end{cases}
$$

En determinadas circunstancias nos puede interesar ocultar el código para mejorar la estética de un cuaderno, especialmente cuando estamos contando conceptos con código (por ejemplo gráficos) y el código -que quizás es demasiado largo- no aporta contenido relevante para el lector.

Si clicas en 'Mostrar código' se desplegará el código oculto. Si quieres volver al estado inicial, basta clicar dos veces sobre la celda-formulario.


````python
#@title <b><font color="steelblue"> Distribución Binomial{ run: "auto" }

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
n =  30#@param {type:"integer"}
p = 0.2 #@param {type:"slider", min:0, max:1, step:0.01}
fig, ax = plt.subplots(figsize=(7, 4))
s = n

x = np.arange(s)
y = binom.pmf(x,n,p)

ax.bar(x, y, color="k", label="Binomial n, p = ({}, {})".format(n,p))
ax.set_xlabel("Number of Trials")
ax.set_ylabel("Probability")
ax.set_xticks(x)
ax.legend()

plt.show()
````
